```python
# read in IPUMS Data and take only full-time full-year workers 

TextFileReader = pd.read_csv("data/usa_00007.csv", chunksize = 2000)

dfList = []
counter = 0
for df in TextFileReader: 
    
    # extract important rows (RIGHT NOW ONLY LOOKING AT FULL TIME WORKERS) 
    #df = df[(df['UHRSWORK'] >= 35) & (df['WKSWORK1'] >= 35)]
    df = df[(df['UHRSWORK'] >= 35)]
    
    dfList.append(df)
    
    # verbose...
    counter += 1 
    if (counter % 100): 
        print("processed ", counter*2000, " rows")

df = pd.concat(dfList, sort=False)
df
```

```python
# Read in Crosswalk Data
import pyreadstat
import requests

# get crosswalk data from url 
url = 'https://s3.us-west-2.amazonaws.com/secure.notion-static.com/34a83796-e75a-4dd9-b74b-a61db985d027/onet_to_occ1990dd.dta?X-Amz-Algorithm=AWS4-HMAC-SHA256&X-Amz-Credential=AKIAT73L2G45O3KS52Y5%2F20210421%2Fus-west-2%2Fs3%2Faws4_request&X-Amz-Date=20210421T050356Z&X-Amz-Expires=86400&X-Amz-Signature=4940ce0f14d28f2620ef67420f4096ff0f9c2107a6491fa49b5d935a56d78832&X-Amz-SignedHeaders=host&response-content-disposition=filename%20%3D%22onet_to_occ1990dd.dta%22'
r = requests.get(url, allow_redirects=True)

open('data/crosswalk.dta', 'wb').write(r.content)

crosswalk, meta = pyreadstat.read_dta('data/crosswalk.dta')
crosswalk
```

```python
crosswalk = crosswalk.groupby('occ').agg('first').reset_index()
crosswalk
```

```python
# Create CPI99 Dictionary 
# SOURCE: https://usa.ipums.org/usa/cpi99.shtml
cpi99 = {} 
cpi99[1980] = 2.295 
cpi99[1990] = 1.344
cpi99[2000] = 0.967
cpi99[2001] = 0.941
cpi99[2002] = 0.926
cpi99[2003] = 0.905
cpi99[2004] = 0.882
cpi99[2005] = 0.853
cpi99[2006] = 0.826
cpi99[2007] = 0.804
cpi99[2008] = 0.774
cpi99[2009] = 0.777
cpi99[2010] = 0.764
cpi99[2011] = 0.741
cpi99[2012] = 0.726
cpi99[2013] = 0.715
cpi99[2014] = 0.704
cpi99[2015] = 0.703
cpi99[2016] = 0.694
cpi99[2017] = 0.679
cpi99[2018] = 0.663
cpi99[2019] = 0.652
```

```python
#df.columns
df = df[['YEAR', 'STATEICP', 'SEX', 'AGE', 'RACE', 'RACED', 'EDUC', 'EDUCD', 'EMPSTAT', 'EMPSTATD', 
         'LABFORCE', 'OCC', 'OCC1990', 'OCC2010', 'OCCSOC', 'IND', 'IND1990', 'WKSWORK1', 'UHRSWORK', 
         'INCWAGE']]
df
```

```python
# group by occupation and industry. 
# Then compute differences between 1980 and 2010 for each individual 
df_1980 = df.loc[df['YEAR'] == 1980]
df_2010 = df.loc[df['YEAR'] == 2010]
```

```python
# calculate real wages in 1999 dollars
df_1980['INCWAGE_REAL'] = df_1980['INCWAGE']*cpi99[1980]
df_2010['INCWAGE_REAL'] = df_2010['INCWAGE']*cpi99[2010]
```

```python
# group by occupation and industry and calculate means for wages
grp_80 = df_1980.groupby(['OCC2010', 'IND1990']).agg('mean').reset_index()
grp_10 = df_2010.groupby(['OCC2010', 'IND1990']).agg('mean').reset_index()
```

```python
# merge the two dataframes with an inner join 
grp_80.columns = [str(x)+'_80' if (x != 'OCC2010') & (x != 'IND1990') else x for x in grp_80.columns]
grp_10.columns = [str(x)+'_10' if (x != 'OCC2010') & (x != 'IND1990') else x for x in grp_10.columns]

df_merged = pd.merge(grp_80, grp_10, how = 'inner', on=['OCC2010', 'IND1990'])
df_merged
```

```python
# Convert OCC2010 to the David Dorne OCC Code 
titles = [] 
dd_codes = [] 

for idx, row in df_merged.iterrows(): 
    #print(row)
    occ_2010 = row['OCC2010']
    
    crow = crosswalk[crosswalk['occ'] == occ_2010]
    
    if (crow.empty): 
        title = None
        dd = None 
    
    else: 
        title = crow['occ1990dd_title'].values[0]
        dd = crow['occ1990dd'].values[0]
    
    titles.append(title)
    dd_codes.append(dd)

df_merged['OCC1990DD'] = dd_codes
df_merged['OCC1990DD_TITLE'] = titles
```

```python
df_merged
```

```python
df_merged = df_merged[df_merged['OCC1990DD'].notnull()]
df_merged
```

```python
# calculate the differences and the log differences
df_merged['DIFF'] = n.abs(df_merged['INCWAGE_REAL_10'] - df_merged['INCWAGE_REAL_80'])
df_merged['LOG_DIFF'] = n.log(df_merged['DIFF'])
```

```python
wage_df = df_merged[['OCC1990DD_TITLE', 'OCC1990DD', 'IND1990', 'INCWAGE_80', 'INCWAGE_10', 'INCWAGE_REAL_80', 
                     'INCWAGE_REAL_10', 'DIFF', 'LOG_DIFF']].sort_values(by='LOG_DIFF')
wage_df
```

```python
wage_df = wage_df.dropna()
wage_df['OCC1990DD'] = wage_df['OCC1990DD'].astype(int)
wage_df
```

```python
# attach exposure scores
exposure_scores = pd.read_csv('data/exposure_by_occ1990dd_lswt2010.csv')
exposure_scores = exposure_scores[['occ1990dd', 'lswt2010', 'pct_software', 'pct_robot', 'pct_ai']]

exposure_df = pd.merge(wage_df, exposure_scores, how='inner', left_on='OCC1990DD', right_on='occ1990dd')
exposure_df
```

```python
# RUN A BASIC REGRESSION
import statsmodels.api as smf

# CREATE INDUSTRY FIXED EFFECTS 
industry_dummies = pd.get_dummies(exposure_df['IND1990']).rename(columns = lambda x: 'IND1990_' + str(x)).iloc[:, 1:]

X = pd.concat([exposure_df['pct_ai'], industry_dummies], axis=1)
Y = exposure_df['LOG_DIFF']
res1 = smf.OLS(Y, X).fit()
print(res1.summary())
```

**REGRESSION FOR EMPLOYMENT**

```python
# group by occupation and industry. 
# Then compute differences between 1980 and 2010 for each individual 
df_1980 = df.loc[df['YEAR'] == 1980]
df_2010 = df.loc[df['YEAR'] == 2010]

# group by occupation and industry and calculate means for wages
emp_grp_80 = df_1980.groupby(['OCC2010', 'IND1990']).agg('sum').reset_index()[['OCC2010', 'IND1990', 'EMPSTAT']]
emp_grp_10 = df_2010.groupby(['OCC2010', 'IND1990']).agg('sum').reset_index()[['OCC2010', 'IND1990', 'EMPSTAT']]
```

```python
emp_grp_80
```

```python
# merge the two dataframes with an inner join 
emp_grp_80.columns = [str(x)+'_80' if (x != 'OCC2010') & (x != 'IND1990') else x for x in emp_grp_80.columns]
emp_grp_10.columns = [str(x)+'_10' if (x != 'OCC2010') & (x != 'IND1990') else x for x in emp_grp_10.columns]

emp_df_merged = pd.merge(emp_grp_80, emp_grp_10, how = 'inner', on=['OCC2010', 'IND1990'])
emp_df_merged
```

```python
def dhs(x, y): 
    return (x-y)/n.mean([x,y])

emp_df_merged['DHS'] = emp_df_merged.apply(lambda x: dhs(x['EMPSTAT_10'], x['EMPSTAT_80']), axis=1)
emp_df_merged
```

```python
# Convert OCC2010 to the David Dorne OCC Code 
titles = [] 
dd_codes = [] 

for idx, row in emp_df_merged.iterrows(): 
    #print(row)
    occ_2010 = row['OCC2010']
    
    crow = crosswalk[crosswalk['occ'] == occ_2010]
    
    if (crow.empty): 
        title = None
        dd = None 
    
    else: 
        title = crow['occ1990dd_title'].values[0]
        dd = crow['occ1990dd'].values[0]
    
    titles.append(title)
    dd_codes.append(dd)

emp_df_merged['OCC1990DD'] = dd_codes
emp_df_merged['OCC1990DD_TITLE'] = titles
emp_df_merged = emp_df_merged[emp_df_merged['OCC1990DD'].notnull()]
emp_df_merged
```

```python
emp_df_merged['OCC1990DD'] = emp_df_merged['OCC1990DD'].astype(int)
exposure_df_emp = pd.merge(emp_df_merged, exposure_scores, how='inner', left_on='OCC1990DD', right_on='occ1990dd')
exposure_df_emp
```

```python
# RUN A BASIC REGRESSION
import statsmodels.api as smf

# CREATE INDUSTRY FIXED EFFECTS 
industry_dummies = pd.get_dummies(exposure_df_emp['IND1990']).rename(columns = lambda x: 'IND1990_' + str(x)).iloc[:, 1:]

X = pd.concat([exposure_df_emp['pct_ai'], industry_dummies], axis=1)
Y = exposure_df_emp['DHS']
res2 = smf.OLS(Y, X).fit()
print(res2.summary())
```
